```python
import pandas as pd
import numpy as np
from efficient_apriori import apriori

#读取数据
dataset =  pd.read_csv('E:/private/python/数据分析课程/L4/L4/MarketBasket/Market_Basket_Optimisation.csv',
                       header =  None)
dataset
dataset1 = dataset.fillna('')
dataset1
#dataset[0][1]
```

```python
#数据处理
index1 = []
for  i in range(7501):
    index1.append(i)
    
temp_pd = pd.DataFrame({},index = index1, columns = [0])
temp_pd
temp_pd = temp_pd.fillna('')
temp_pd
```

```python
#数据处理
for i in range(7501):
    for j in range(20):
        if dataset1[j][i] != '':
            temp_pd.iloc[i,0] = temp_pd.iloc[i,0] + '|' + dataset1[j][i]
temp_pd
```

```python
#编码
hot_enoded = temp_pd[0].str.get_dummies('|')
hot_enoded
```

```python
#数据挖掘
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
frequent_itemsets = apriori(hot_enoded, min_support=0.05, use_colnames=True)
frequent_itemsets
frequent_itemsets = frequent_itemsets.sort_values(by="support" , ascending=False) 
frequent_itemsets
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```
